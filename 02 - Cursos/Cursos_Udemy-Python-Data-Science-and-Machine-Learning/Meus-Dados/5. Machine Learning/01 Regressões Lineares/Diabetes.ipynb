```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split,cross_val_predict
from sklearn import metrics
%matplotlib inline
sns.set_palette("RdBu_r")
sns.set_style('darkgrid')
print('Importações OK!')
```

Carregamento dos dados

```python
diabetes = datasets.load_diabetes()
X, y = diabetes.data, diabetes.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
lm = linear_model.LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
predicts = lm.predict(X_test)
```

```python
print('coeficientes: \n', lm.coef_)
```

```python
plt.scatter(y_test,predicts)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
print('MAE: %.2f' % metrics.mean_absolute_error(y_test, predicts))
print('MSE: %.2f' % metrics.mean_squared_error(y_test, predicts))
print('RMSE: %.2f' % np.sqrt(metrics.mean_squared_error(y_test, predicts)))
rmse = np.sqrt(metrics.mean_squared_error(y_test, predicts))
```

```python
[y_test.min()+rmse, y_test.max()+rmse], [y_test.min()-rmse, y_test.max()-rmse]
```

```python
plt.scatter(y_test,predicts)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.plot([predicts.min(), predicts.max()], [predicts.min()+rmse, predicts.max() + rmse], 'k--', lw=0.5)
plt.plot([predicts.min() + rmse, predicts.max()+ rmse], [predicts.min(), predicts.max()], 'k--', lw=0.5)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
plt.show()
```

```python
print('Variance score: %.2f' % metrics.r2_score(y_test, predicts))
print('se der 1, nosso modelo está maravilhosamente bom')
```

```python
sns.histplot((y_test-predicts),bins=50, kde=True);
```

```python
lr = linear_model.LinearRegression()
X, y = datasets.load_diabetes(return_X_y=True)

# cross_val_predict returns an array of the same size as `y` where each entry
# is a prediction obtained by cross validation:
predicted = cross_val_predict(lr, X, y, cv=10)

fig, ax = plt.subplots()
ax.scatter(y, predicted, edgecolors=(0, 0, 0))
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```

```python
print('Variance score: %.2f' % metrics.r2_score(y, predicted))
print('se der 1, nosso modelo está maravilhosamente bom')
```
