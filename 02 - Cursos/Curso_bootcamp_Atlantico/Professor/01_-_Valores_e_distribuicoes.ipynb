# Valores e distribuições

## Gráficos de barra
 - Barras retangulares e comprimento proporcional aos valores que ele apresenta
 - Vertical ou na horizontal
 - Ilustrar comparações entre variáveis quantitativas

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame({
    "arrecadacao": [71565498, 36169328, 19928525, 8805843, 7316746],
    "titulo": [
        "Star Wars: The Last Jedi",
        "Jumanji: Welcome to the Jungle",
        "Pitch Perfect 3",
        "The Greatest Showman",
        "Ferdinand"
    ]
}).sort_values(by='arrecadacao')
df
```

```python
df.plot.bar(x="titulo", y="arrecadacao")
```

```python
plt.figure(figsize=(10, 5))
plt.barh(df.titulo, df.arrecadacao)
plt.xlabel("Arrecadação do fim de semana (USD)")
plt.ylabel("Título")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(
    data=df.sort_values(by='arrecadacao', ascending=False),
    x='arrecadacao', y='titulo', orient='h', color='orange'
)
plt.xlabel("Arrecadação do fim de semana (USD)")
plt.ylabel("Título")
plt.show()
```

## Gráficos de densidade
 - Visualizar a distribuição de probabilidade subjacente dos dados
 - Estimativa de densidade de kernel (kernel density estimation, KDE). 
 - Curva contínua (o kernel) com uma pequena largura
 - kernel mais usado é um kernel gaussiano

```python
sns.kdeplot(data=df_alturas, x='alturas')
plt.xlabel("Alturas (cm)")
plt.ylabel("Quantidade")
plt.show()
```

```python
sns.histplot(data=df_alturas, x='alturas', bins=30, kde=True)
plt.xlabel("Alturas (cm)")
plt.ylabel("Quantidade")
plt.show()
```

## Boxplot

```python
sns.boxplot(x=df_alturas["alturas"])
plt.show()
```
