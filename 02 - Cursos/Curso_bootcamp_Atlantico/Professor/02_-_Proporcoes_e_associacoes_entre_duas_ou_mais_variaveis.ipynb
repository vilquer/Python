# Proporções e associações entre duas ou mais variáveis
## Gráfico de pizza
 - Gráficos divididos em setores, cada setor da pizza exibe o tamanho de uma parte da informação
 - Visualiza claramente os dados como proporções de um todo
 - Enfatiza visualmente frações simples

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid")
```

```python
df = pd.DataFrame({
    "value": np.random.randint(low=30, high=500, size=5),
    "group": [f"Group {i}" for i in range(1, 6)]
}).sort_values(by="value", ascending=False)
df
```

```python
colors = sns.color_palette('pastel')[0:5]

#create pie chart
plt.pie(df.value, labels=df.group, colors=colors, autopct='%.0f%%')
plt.show()
```

## Gráficos de dispersão
 - Representações gráficas do relacionamento entre duas variáveis numéricas
 - Utiliza pontos para representar essa relação
 - Cada ponto representa o valor de uma variável no eixo horizontal e o valor de outra variável no eixo vertical

```python
tips = sns.load_dataset("tips")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", alpha=.3) # 
plt.show() # conta x gorjeta
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
plt.show() # conta x gorjeta + horário
```

```python
sns.scatterplot(
    data=tips, x="total_bill", y="tip", hue="size", size="size",
    sizes=(20, 200), hue_norm=(0, 7), legend="full"
)
plt.show() # conta x gorjeta + tamanho da mesa
```

```python
sns.kdeplot(
    data=tips, x="total_bill", y="tip", fill=True
)
```
