## .loc (localização) vs .iloc (localização indexada)

Como .loc e .iloc são maneiras de **acessar** os dados do dataframe ele é seguido de **[ ]** em vez de ().

````python
df.iloc[0]
````

````python
df.iloc[:3]
````

````python
df.loc[0]
````

````python
df.loc[:3]
````

O .loc tem usa o index para localizar a posição que estamos tentando acessar.
Já o .iloc, cria uma interface de indexação numérica (de zero até tamanho do dataframe) para acessar as posições.

````python
df = df.set_index('datas')
df
````

````python
df.index
````

````python
df.iloc[0]
````

````python
df.loc['2019-01-01']
````

Como o iloc cria uma interface de indexação, ele acaba criando uma interface dos dados também

````python
df.iloc[:10]['valores'] = 100
df
````

````python
df.loc[dates[:10], 'valores'] = 10
df
````

Para o .loc

* a primeira posição é o index ou array de indexes para acessar as localizações (linhas)
* a segunda posição é a coluna ou a lista de colunas desejamos ver/alterar

Para o iloc

* a primeira posição é o index ou array de indexes para acessar as localizações (linhas)
* a segunda posição é o index das colunas que desejamos ver

````python
df.loc['2018-10-05', 'taxas']
````

## Dropando Duplicadas

.drop_duplicates tem 3 argumentos:

* keep: 'first', 'last', False (default='first')
* subset: coluna ou lista de colunas (default='None')
* inplace: False (default='False')

````python
data = {
    'nomes':['Augusta da Silva', 'Angústia Tássida',
             'Maurio Silva', 'Critéria Elevada',
             'Cândido Diase de Silva'],
    'idades': [28, 38, 15, 64, 91],
    'cidade': ['Boqueirão do Leão', 'Palmitinho', 'Faxinal do Soturno', 'Pinto Bandeira', 'Faxinalzinho']
}
df = pd.DataFrame(data)
df
````

````python
duplicado = df.append(df, ignore_index=True)
duplicado
````

>O operador .merge executa o cruzamento levando em consideração algum valor da tabela de referência que exista na tabela adicional.
>
>Ele considera a tabela referência (de onde o método será chamado) o "left" e a adicional (parâmetro do método) o "right"

````python
df1.merge(df2, on='A')
````

> Se o nome entre a nossa referência de merge for diferente teremos que usar a nossa nomenclatura right e left

````python
df1.merge(df2, left_on='B', right_on='Bs')
````

> Podemos trocar a nomenclatura das colunas repetidas se desejarmos

````python
df1.merge(df2, left_on='B', right_on='Bs', suffixes=('', '_novo'),)
````

## Ólia quim xôu

````python
def leitor_de_csv(texto):
    linhas = texto.split('\n')
    
    primeira_linha = linhas[0]
    linhas = linhas[1:]
    colunas = primeira_linha.split(',')
    
    csv = {}
    for coluna in colunas:
        csv[coluna] = []
    
    for linha in linhas:
        valores = linha.split(',')
        
        if len(valores) != len(colunas):
            continue
        
        for i, valor in enumerate(valores):
            coluna_para_adicionar = colunas[i]
            csv[coluna_para_adicionar].append(valor)
    return csv
````

````python
with open('teste.csv', 'r') as f:
    texto = f.read()
````

````python
print(texto)
````

````python
data = leitor_de_csv(texto)
data
````
