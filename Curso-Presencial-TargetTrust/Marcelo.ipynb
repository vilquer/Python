```python
import pandas as pd
from pandas.tseries.offsets import MonthEnd
```

```python
data = {
    'funcionario': ['Marcelo', 'Predo', 'Varzeano', 'Exatonildo'],
    'data_inicio': ['25/03/2020', '22/02/2020', '22/02/2020', '21/01/2020'],
    'data_termino': ['05/04/2020', '10/03/2020', '21/04/2020', '31/01/2020'],
}
df = pd.DataFrame(data)
df['data_inicio'] = pd.to_datetime(df['data_inicio'], dayfirst=True)
df['data_termino'] = pd.to_datetime(df['data_termino'], dayfirst=True)
df
```

```python
df['duracao'] = df['data_termino'] - df['data_inicio']
df
```

```python
splited_df = pd.DataFrame(columns=['funcionario', 'data_inicio', 'data_termino'])
```

```python
splited_df
```

```python
splited_df = pd.DataFrame(columns=['funcionario', 'data_inicio', 'data_termino'])
i = 0

for _, row in df.iterrows():
    worker = row['funcionario']
    beginning = row['data_inicio']
    end = row['data_termino']
    
    iterations = end.month - beginning.month
    
    for _ in range(iterations + 1):
        limit = beginning + MonthEnd(1)
        
        if limit >= end:
            splited_df.loc[i] = worker, beginning, end           
        else:
            splited_df.loc[i] = worker, beginning, limit
            beginning = limit
        i += 1

splited_df['duracao'] = splited_df['data_termino'] - splited_df['data_inicio']
splited_df
```
