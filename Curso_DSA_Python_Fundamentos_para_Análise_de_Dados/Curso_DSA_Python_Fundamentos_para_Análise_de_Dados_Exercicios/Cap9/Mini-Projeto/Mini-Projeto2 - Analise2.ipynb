```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
list(df)
```

## Número de veículos pertencentes a cada marca

```python
# Crie um Plot que mostre o número de veículos pertencentes a cada marca
sns.set_style("whitegrid")
g = sns.catplot(y='brand', kind="count", data=df, palette='Reds_r', height=14, aspect=1)
g.ax.set_title("Veículos Por Marca",fontdict={'size':20})
g.ax.xaxis.set_label_text("Número de Veículos", fontdict= {'size':20})
g.ax.yaxis.set_label_text("Marca", fontdict= {'size':20})
```

```python
# Salvando o plot
g.savefig(("plots/Analise2/brand-vehicleCount.png"))
```

## Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```python
# Crie um Plot com o Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio
```

```python
fig, ax = plt.subplots(figsize=(20,10))
sns.barplot(x='vehicleType', y='price', data=df, hue='gearbox', palette='Blues')
ax.set_title('Preço médio dos veículos por tipo de veículo e caixa de câmbio', fontdict={'size':20})
ax.set_xlabel('Tipo de veículo', fontdict={'size':20})
ax.set_ylabel('Preço médio', fontdict={'size':20})
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise2/vehicletype-gearbox-price.png")
```
