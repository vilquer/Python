## Análise 1

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
list(df)
```

## Distribuição de Veículos com base no Ano de Registro

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro
fig, ax = plt.subplots(figsize=(16,8))
sns.distplot(df["yearOfRegistration"], color="#0554F2",kde=True, ax=ax)
plt.title('Distribuição de Veículos com base no Ano de Registro', fontsize= 20, loc='center')
plt.ylabel('KDE',fontsize= 20)
plt.xlabel('Ano do registro',fontsize= 20)
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/vehicle-distribution.png")
```

## Variação da faixa de preço pelo tipo de veículo

```python
# Crie um Boxplot para avaliar os outliers
fig, ax = plt.subplots(figsize=(16,8))
sns.set(style="whitegrid")
ax = sns.boxplot(x='vehicleType',y='price', data= df)
plt.title('Variação de preços por tipo de veículo', fontsize= 20, loc='center')
plt.ylabel('Preço',fontsize= 20)
plt.xlabel('Tipo de veículo',fontsize= 20)
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/price-vehicleType-boxplot.png")
```

## Contagem total de veículos à venda conforme o tipo de veículo

```python
# Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria 
sns.set_style('white')
ax = sns.catplot(x='vehicleType', data=df, palette ='Blues', kind="count", height=14, aspect=1)
ax.ax.set_xlabel('Tipo de veículo', fontsize= 20)
ax.ax.set_ylabel('Total de veículos', fontsize= 20)
ax.ax.set_title('Veículos a venda por tipo de veículo', fontdict= {'size':20} )
for p in ax.ax.patches:
        ax.ax.annotate(p.get_height(), (p.get_x()+0.25, p.get_height()+500))
```

```python
# Salvando o plot
ax.savefig("plots/Analise1/count-vehicleType.png")
```
