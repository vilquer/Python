```python
# Imports
import pandas as pd
import numpy as np
```

```python
# Carrega o arquivo
load_file = "dados_compras.json"
purchase_file = pd.read_json(load_file, orient = "records")
purchase_file.head()
```

## Informações Sobre os Consumidores

```python
# Implemente aqui sua solução
tot_cli = purchase_file.Login.count()
display('Número total de consumidores é: ' + str(tot_cli))
```

## Análise Geral de Compras

```python
# Implemente aqui sua solução
#Número de itens exclusivos
mask = ~purchase_file['Item ID'].duplicated()
display('Número de itens exclusivos é: ' + str(purchase_file[mask == True]['Item ID'].count()))

#Preço médio de compra
display('Preço médio de compra: ' + str(format(purchase_file['Valor'].mean(), '.2f')))

#Número total de compras
display('Número total de compras é: ' + str(purchase_file.Valor.count()))
#Rendimento total
display('Rendimento total é: ' + str(purchase_file.Valor.sum()))
```

## Análise Demográfica

```python
# Implemente aqui sua solução
#Porcentagem e contagem de compradores masculinos
tot_masc = purchase_file.Sexo[purchase_file['Sexo'] == 'Masculino'].count()
display('O Total de masculino é: ' + str(tot_masc))
perc_masc = tot_masc / tot_cli * 100
display('O Percentual masculino é: ' + str(format(perc_masc, '.2f')))
#Porcentagem e contagem de compradores do sexo feminino
tot_fem = purchase_file.Sexo[purchase_file['Sexo'] == 'Feminino'].count()
display('O Total de feminino é: ' + str(tot_fem))
perc_fem = tot_fem / tot_cli * 100
display('O Percentual feminino é: ' + str(format(perc_fem, '.2f')))
#Porcentagem e contagem de outros / não divulgados
tot_out = purchase_file.Sexo[purchase_file['Sexo'] == 'Outro / Não Divulgado'].count()
display('O Total de  é: ' + str(tot_out))
perc_out = tot_out / tot_cli * 100
display('O Percentual  é: ' + str(format(perc_out, '.2f')))
```

## Informações Demográficas Por Gênero

```python
# Implemente aqui sua solução
purchase_file[['Sexo','Login']].groupby(['Sexo']).count().reset_index()
```

## Análise de Compras Por Gênero

```python
# Implemente aqui sua solução
#Número de compras
display('Numero de compras por gênero')
display(purchase_file[['Sexo','Login']].groupby(['Sexo']).count().rename(columns={'Login': 'Quantidade'}).reset_index())
#Preço médio de compra
display('Preço médio das compras por gênero')
display(purchase_file[['Sexo','Valor']].groupby(['Sexo']).mean().rename(columns={'Valor': 'Média'}).reset_index())
#Valor Total de Compra
display('Total das compras por gênero')
display(purchase_file[['Sexo','Valor']].groupby(['Sexo']).sum().rename(columns={'Valor': 'Total'}).reset_index())
#Compras for faixa etária
display('Total das compras por Idade')
display(purchase_file[['Idade','Valor']].groupby(['Idade']).sum().rename(columns={'Valor': 'Total'}).reset_index())

```

## Consumidores Mais Populares (Top 5)

```python
# Implemente aqui sua solução
#** Identifique os 5 principais compradores pelo valor total de compra e, em seguida, liste (em uma tabela): **
#Login
#Número de compras
#Preço médio de compra
#Valor Total de Compra
#Itens mais populares


purchase_file['Soma Valor'] = purchase_file.groupby('Login')['Valor'].transform(np.sum)
purchase_file['Quantidade compras'] = purchase_file.groupby('Login')['Valor'].transform(np.count_nonzero)
purchase_file['Valor médio'] = purchase_file.groupby('Login')['Valor'].transform(np.mean)
purchase_file['Soma Items'] = purchase_file.groupby('Item ID')['Item ID'].transform(np.count_nonzero)
purchase_file.sort_values(by=['Soma Valor'], ascending=False, inplace=True)
df_temp = purchase_file.drop_duplicates(subset=['Login'], keep="first")
df_temp.sort_values(by=['Soma Valor'], ascending=False, inplace=True)
df_temp[['Login','Quantidade compras', 'Valor médio', 'Valor', 'Item ID']].head(5)
```
