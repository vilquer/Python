```python
# Bibliotecas
import matplotlib.pyplot as plt

# Criandos os eixos e o nome
x = [0, 1, 2, 3, 4]
y = [5, 15, 8, 22, 36]
nomes = ('A', 'B', 'C', 'D', 'E')

# Criando as barras
plt.barh(x, y)

# colocando os nomes no eixo x
plt.yticks(x, nomes)

# Melhorando a aparência
plt.style.use('fivethirtyeight')#
plt.title('Grafico teste')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')

# mostrando o gráfico
plt.show()
```
