```python
# Bbliotecas
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Lista de Palavras
texto=("""Lorem ipsum dolor sit amet, consectetur adipiscing elit. 
Maecenas viverra in urna eu tempor. Curabitur pulvinar sapien vitae auctor accumsan. 
Nunc congue gravida mi, eu sodales sem elementum fringilla. 
Mauris tellus ipsum, hendrerit commodo massa at, mattis venenatis nisi.""")

# criando um objeto wordcloud 
nuvem_palavras = WordCloud(width=480, 
                           height=480, 
                           background_color="gray",
                           colormap="Blues",
                           margin=1).generate(texto)

# Mostrando a imagem gerada:
plt.figure(figsize= (6,6))
plt.imshow(nuvem_palavras, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
