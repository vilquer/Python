```python
# Números inteiros
x = 1
y = 2
resultado = x + y
print(resultado)
print(type(resultado))
```

```python
# Números de ponto flutuante
x = 1.1
y = 2.2
resultado = x + y
print(resultado)
print(type(resultado))
```

```python
# Números compéxos
x = 1+2j
y = complex(3,4)
resultado = x + y
print(resultado)
print(type(resultado))
```

```python
# Strings
str_vazia = ''
letra = 'a'
Frase = 'olá mundo'
print(type(str_vazia))
print(type(letra))
print(type(Frase))
```

```python
# Três formas de fazer a mesma coisa
var_1 = 'Três '
var_2 = "formas de "
var_3 = '''fazer a '''
var_4 = """mesma coisa! """
print(var_1 + var_2 + var_3 + var_4)
```

```python
# Variáveis booleanas (verdadeiro/falso)
x = True
y = False
resultado = x == y # = Atribuição, == Comparação
print(resultado)
print(type(resultado))
```
