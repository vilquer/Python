```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
list(df)
```

```python
# Calcule a média de preço por marca e por veículo
```

```python
df['brand'].unique()
```

```python
df_agrupado = df.groupby(['brand','vehicleType'])['price'].mean().reset_index(drop=False)
```

```python
df_temp = pd.DataFrame()
df_temp['AVGPrice'] = df_agrupado["price"].astype(int)
df_temp['brand'] = df_agrupado["brand"]
df_temp['vehicleType'] = df_agrupado["vehicleType"]
```

```python
df_temp[df_temp['brand'] == 'volkswagen']
```

```python
tri = df_agrupado.pivot("brand","vehicleType", "price")
```

```python
tri.fillna(0,inplace=True)
```

```python
tri.isnull().value_counts()
```

```python
tri
```

## Preço médio de um veículo por marca, bem como tipo de veículo

```python
# Crie um Heatmap com Preço médio de um veículo por marca, bem como tipo de veículo
```

```python
fig, ax = plt.subplots(figsize=(15,20))
sns.heatmap(tri, linewidths=1, cmap="YlGnBu", annot=True, fmt = '.0f', ax=ax)
ax.set_title("Heatmap - Preço médio de um veículo por marca e tipo de veículo",fontdict={'size':20})
ax.xaxis.set_label_text("Tipo de Veículo",fontdict= {'size':20})
ax.yaxis.set_label_text("Marca",fontdict= {'size':20})
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise4/heatmap-price-brand-vehicleType.png")
```
