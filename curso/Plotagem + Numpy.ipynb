# Cores

```python
cores = {
    'b': 'blue',
    'g': 'green',
    'r': 'red',
    'c': 'cyan',
    'm': 'magenta',
    'y': 'yellow',
    'k': 'black',
    'w': 'white'
}
```

```python
plt.figure(figsize=(20, 10))
for i, (cor, nome_cor) in enumerate(cores.items()):
    plt.plot(x, x ** (i / len(cores) - 1), c=cor, label=nome_cor)
plt.legend()
```

# linestyles

```python
linestyles = ['-', '--', '-.', ':']
plt.figure(figsize=(20, 10))
for i, linestyle in enumerate(linestyles):
    plt.plot(x,  x ** (i / len(cores)) + 1, linestyle=linestyle)
```

# Repetir o gráfico do array1 com sua média e limitantes usando cores melhores e outros linestyles.

```python
plt.plot(x, array1, c='b')
plt.axhline(array1.mean() + array1.std(), c='r', linestyle='--')
plt.axhline(array1.mean(), c='k')
plt.axhline(array1.mean() - array1.std(), c='g', linestyle='--')
```

# Entendendo o mecanismo

## Dados

```python
y = np.linspace(0, 5, 501)
x = np.arange(len(y)) 
y_quadratico = y ** 2
y_log = np.log(y)
```

## Primeiro: Só os plots das linhas

```python
plt.plot(y)
plt.plot(y_quadratico)
plt.plot(y_log)
```

## Segundo: x e y labels + plt.show de boas práticas!

```python
plt.xlabel('Passos')
plt.ylabel('Magnitude')
plt.plot(x, y)
plt.plot(x, y_quadratico)
plt.plot(x, y_log)
plt.show()
```

## Terceiro: Mudando o Figsize

```python
plt.figure(figsize=(20, 10))
plt.xlabel('Passos')
plt.ylabel('Magnitude')
plt.plot(x, y)
plt.plot(x, y_quadratico)
plt.plot(x, y_log)
plt.show()
```

## Quarto: Adicionando um título

```python
plt.figure(figsize=(20, 10))
plt.title('Plotinho Querido')

plt.xlabel('Passos')
plt.ylabel('Magnitude')
plt.plot(x, y)
plt.plot(x, y_quadratico)
plt.plot(x, y_log)
plt.show()
```

## Quinto: Legend, grid, xticks

```python
plt.figure(figsize=(20, 10))
plt.title('Plotinho Querido')
plt.xlabel('Passos')
plt.ylabel('Magnitude')
plt.plot(x, y, label='Linear')
plt.plot(x, y_quadratico, label='Quadrática')
plt.plot(x, y_log, label='Log base e')
plt.legend()
plt.grid()
plt.xticks(x[::100], [0, 1, 2, 3, 4, 5])
plt.show()
```

# Taxas + Rendimento

1. criar um array de taxas (entre 0 e 0.05) com 12 valores
1. computar o rendimento para um investimento de 10 000 táoqueis:
    1. Como juros simples
    1. Como juros compostos
1. plotar os rendimentos mensais com:
    1. Tamanho de imagem adequado
    1. Legenda
    1. Cores 
    1. Título
    1. Eixo X e Eixo Y

```python
def calcula_retorno_mensal_simples(taxas, valor_entrada):
    return taxas * valor_entrada

def calcula_rendimento_mensal_composto(taxas, valor_inicial):
    taxas1 = (taxas + 1).cumprod()
    diferenca = np.diff(taxas1)
    taxas_de_rendimento = np.append(taxas[0], diferenca)
    return taxas_de_rendimento * valor_inicial
```

```python
taxas = np.random.rand(12) / 20
x = np.arange(len(taxas))
retorno_juros_simples = calcula_retorno_mensal_simples(taxas, 10000)
retorno_juros_compostos = calcula_rendimento_mensal_composto(taxas, 10000)

plt.figure(figsize=(20, 10))
plt.title('Composto vs Simples')
plt.plot(x, retorno_juros_compostos, label='Retorno Juros Compostos')
plt.plot(x, retorno_juros_simples, label='Retorno Juro Simples')
plt.xlabel('Meses')
plt.ylabel('Rendimento')
plt.legend()
plt.show()
```

# Refazer o gráfico para taxas de 10% ao mês

```python
taxas = np.ones(24) / 10
x = np.arange(len(taxas))
retorno_juros_simples = calcula_retorno_mensal_simples(taxas, 10000)
retorno_juros_compostos = calcula_retorno_mensal_composto(taxas, 10000)

plt.figure(figsize=(20, 10))
plt.title('Composto vs Simples')
plt.plot(x, retorno_juros_compostos, label='Retorno Juros Compostos')
plt.plot(x, retorno_juros_simples, label='Retorno Juro Simples')
plt.xlabel('Meses')
plt.ylabel('Rendimento')
plt.legend()
plt.show()
```

# Bônus 1

```python
produto1_rentabilidades = [40, 67, 82, 10, 22, 39]
produto2_rentabilidades = [20, 27, 73, 41, 33, 61]

x = np.arange(len(produto1_rentabilidades))
width = 0.3

plt.figure(figsize=(16, 6))
plt.title('Mix rentabilidades produto 1 e 2')
plt.bar(x, produto1_rentabilidades, width=width, color='#a3acff', label='Produto 1')
plt.bar(x, produto2_rentabilidades, width=width, color='#f9bc86', label='Produto 2', bottom=produto1_rentabilidades)
plt.xlabel('Mes')
plt.ylabel('Rantabilidade')
plt.xticks(x, ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun'])
plt.show()
```
