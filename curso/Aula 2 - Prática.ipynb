## Prática: usar o rand e o randint

* criar um array de taxas, não permitir que existam valores maiores que 10% (os valores podem alcançar 10, porém não mais que 10)
* criar um array de valores

````python
taxas = np.random.rand(100) / 10
valores = np.random.randint(50, 10000, 100)
````

# Calcular o rendimento da nossa carteira

````python
rendimentos = taxas * valores
````

````python
np.sum(rendimentos)
````

## Encontrar a combinação de taxa e valor que mais trouxe rendimento

````python
index_maior_rendimento = np.argmax(taxas * valores)
print(f'A combinação que gerou maior rendimento foi taxa {100 * taxas[index_maior_rendimento]:.3f}% e valor {valores[index_maior_rendimento]}.')
````

O argmax encontra o index da maior elemento do array. Como taxas, valores e rendimentos estão alinhados poderemos usar os indexes para encontrar os valores equivalentes.

## Foi a maior taxa que gerou o maior valor de rendimento?

````python
index_maior_taxa = np.argmax(taxas)
if index_maior_taxa == index_maior_rendimento:
    print('Sim')
else:
    print('Não')
````

## Foi a menor taxa que gerou o menor valor de rendimento?

````python
index_menor_taxa = np.argmin(taxas)
index_menor_rendimento = np.argmin(rendimentos)
if index_menor_rendimento == index_menor_taxa:
    print('Sim')
else:
    print('Não')
````

## Foi o maior valor que gerou o maior rendimento?

````python
index_maior_valor = np.argmax(valores)
if index_maior_valor == index_maior_rendimento:
    print('Sim')
else:
    print('Não')
````

## Foi o menor valor que gerou o menor rendimento?

````python
index_menor_valor = np.argmax(valores)
if index_menor_valor == index_menor_rendimento:
    print('Sim')
else:
    print('Não')
````

## Encontrar as taxas que são superiores à média + desvio padrão

Para calcular a média usaremos o .mean e para calcular o desvio padrão usaremos o .std

````python
mstd_mask = taxas > (taxas.mean() + taxas.std())
taxas[taxas > taxas.mean() + taxas.std()]
````

Usando o np.where para recuperar os indexes onde a nossa máscara tem valores True:

> O np.where retorna uma lista de indexes, precisaremos acessar a primeira posição para termos os nossos indexes.

````python
indexes_das_maiores_taxas = np.where(taxas > taxas.mean() + taxas.std())[0]
indexes_das_maiores_taxas
````

## Usando o indexes calculado, calcular quantos porcento do somatório de nossa carteira vem das "maiores" taxas

````python
valores_taxas_altas = np.sum(taxas[indexes_das_maiores_taxas] * valores[indexes_das_maiores_taxas])
rendimento_carteira = np.sum(rendimentos)
representatividade_das_taxas = valores_taxas_altas / rendimento_carteira

print(f'As taxas altas representam {(100 * representatividade_das_taxas):.4f}% do rendimento de nossa carteira.')
````

## Teste

Entendemos o cálculo, porém com estes valores aleatórios não conseguimos saber se fizemos todas as operações de maneira correta. Vamos testar usando taxas que conhecemos como 10%...

````python
teste = np.ones(100) / 10
teste
````

````python
teste1 = (teste + 1).cumprod()
````

````python
np.append(teste[0], np.diff(teste1)) * 10000
````

Agora que executamos o batimento a cima, podemos fazer a função que executa o cálculo do rendimento mensal para garantir que sempre executaremos as operações na ordem certa!

````python
def calcula_rendimento_mensal(taxas, valor_inicial):
    taxas1 = (taxas + 1).cumprod()
    diferenca = np.diff(taxas1)
    taxas_de_rendimento = np.append(taxas[0], diferenca)
    return taxas_de_rendimento * valor_inicial
````

````python
calcula_rendimento_mensal(teste, 10000)
````

````python
calcula_rendimento_mensal(taxas, 10000)
````
