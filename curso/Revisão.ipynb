# Criar 3 arrays de 100 elementos:

1. array1: um range
1. array2: com números aleatórios entre -0.25 e 0.25
1. array3: com números aleatórios entre 1, 100

````python
array1 = np.arange(100)
array2 = (np.random.rand(100) - 0.5) / 2
array3 = np.random.randint(1, 100, 100)
````

# Verificar se:

1. todos os elementos em array1 são maiores que array3
2. se existem valores em array3 que vezes array2 são maiores que array1

## 1

````python
if np.all(array1 > array3):
    print('Sim')
else:
    print('Não')
````

## 2

````python
if np.any(array3 * array2 > array1):
    print('Sim')
else:
    print('Não')
````

# Acessar os valores de array1 que são maiores que array3

````python
array1[array1 > array3]
````

# Existe algum elemento em que array1 * array2 é maior que a média de array3 + um desvio padrão? 

````python
if np.any((array1 * array2) > (array3.mean() + array3.std())):
    print('Sim')
else:
    print('Não')
````

# E o contrário?

````python
if np.any((array1 * array2) <= (array3.mean() + array3.std())):
    print('Sim')
else:
    print('Não')
````

# plotar array1, array2 e array3 com legenda

````python
plt.figure(figsize=(8, 6))
plt.plot(array1, label='Array 1')
plt.plot(array2, label='Array 2')
plt.plot(array3, label='Array 3')
plt.legend()
plt.show()
````

# Criar um dataframe a partir do dicionário que tenha o seguinte conjunto de chave: valor

* 100 taxas: taxas aleatórias entre -0.025 e 0.025
* 100 dias: range de datas variando 1 dia.


````python
data = {
    'taxas': (np.random.rand(100) - 0.5) / 20,
    'dias': pd.date_range(start='2019/01/22', periods=100)
}
````

````python
df = pd.DataFrame(data)
df
````

# Adicionar o dataframe o cumprod e o cumsum dessas taxas

````python
df['cumprod'] = (df['taxas'] + 1).cumprod()
df['cumsum'] = df['taxas'].cumsum() + 1
````

````python
df
````

# criar o interact que me permite selecionar os dias com um slider

# Bônus: Carregando dados de um SQL-Like

## Queries para pegar a primeira linha e ter acesso aos nomes das colunas da tabela. (Não esquecer de trocar "table_name" pelo nome da sua tabela!)

````python

query_oracle = """
SELECT *
FROM table_name
WHERE
    ROWNUM = 1;
"""
print(query_oracle)
````

````python
query_sql_server = """
SELECT TOP 1 *
FROM table_name
"""
print(query_sql_server)
````

````python
query_postgres = """
SELECT *
FROM table_name
LIMIT 1
"""
print(query_postgres)
````

## Imports necessários

````python
import pyodbc  # SQL-Server
import psycopg2  # PostgreSQL
import cx_Oracle  # OracleDB
````

## Objetos de conexão

````python
ss_conn = pyodbc.connect()
pg_conn = psycopg2.connect()
oracle_conn = cx_Oracle.connect()
````

````python
ss_df = pd.read_sql_query(query_sql_server, ss_conn)
pg_df = pd.read_sql_query(query_postgres, pg_conn)
oracle_df = pd.read_sql_qyery(query_oracle, oracle_conn)
````

## chunksize

é um parâmetro do read_sql_query. quando é fornecido nosso read_sql_query passa a fornecer os dados em subdataframes de chunksize linhas. O funcionamento é MUITO semelhante ao do read_csv, com chunksize.
